```python
# Importar algumas bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Carregar a base sinasc_RO_2019.csv

sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.head(10)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Por serem variáveis de tipos diferentes criarei um gráfico para cada variável

plt.close('all')
plt.rc('figure', figsize=(12,25))
fig, axes = plt.subplots(5,1)

sns.countplot(ax = axes[0], x='APGAR1', data=sinasc)
sns.countplot(ax = axes[1], x='APGAR5', data=sinasc)
sns.countplot(ax = axes[2], x='SEXO', data=sinasc)
sns.countplot(ax = axes[3], x='GESTACAO', data=sinasc)
graf_est = sns.countplot(ax = axes[4], x='ESTCIVMAE', data=sinasc)
labels_est = graf_est.set_xticklabels(['Solteira', 'Casada', 'Viuva', 'Separado Judicialmente/Divorciada', 'União Consensual', 'Ignorado'],
                            rotation=30, fontsize='small')
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10,10))

graf = sns.barplot(x='CONSULTAS', y='APGAR5', data=sinasc)
labels = graf.set_xticklabels(['nenhuma', 'de 1 a 3', 'de 4 a 6', '7 e mais', 'Ignorado'],
                            rotation=30, fontsize='small')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(24, 12))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0,0], x='SEXO', data=sinasc)
sns.countplot(ax = axes[0,1], x='GESTACAO', data=sinasc)
graf_est = sns.countplot(ax = axes[1,0], x='ESTCIVMAE', data=sinasc)
labels_est = graf_est.set_xticklabels(['Solteira', 'Casada', 'Viuva', 'Separado Judicialmente/Divorciada', 'União Consensual', 'Ignorado'],
                            rotation=30, fontsize='small')
graf_cons = sns.countplot(ax = axes[1,1], x='CONSULTAS', data=sinasc)
labels_cons = graf_cons.set_xticklabels(['nenhuma', 'de 1 a 3', 'de 4 a 6', '7 e mais', 'Ignorado'])
                            
```
