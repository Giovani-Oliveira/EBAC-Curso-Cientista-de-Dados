1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
sns.set(rc={'figure.figsize':(12,7)})
ax = sns.countplot(x='data_ref',  hue = 'sexo', data=renda_raw)

tick_labs = renda_raw['data_ref'].unique()

ticks = ax.set_xticks(list(range(renda_raw['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)
title = ax.set_title('Sexo')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````
