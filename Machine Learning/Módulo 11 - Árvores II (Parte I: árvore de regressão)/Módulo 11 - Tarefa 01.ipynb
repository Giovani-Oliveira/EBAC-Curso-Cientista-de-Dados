# Árvores de regressão - exercícios 01

Nesta bateria de exercícios, vamos botar em prática o que aprendemos com a base de dados imobiliários de Boston, que pode ser baixada do módulo ```datasets``` do scikitlearn. Essa base de dados possui uma linha por cidade, o objetivo é prever, ou fornecer uma estimativa de ```MEDV```, ou o valor mediano de casas ocupadas pelo dono, dadas as demais variáveis usadas como explicativas.

A descrição das variáveis está abaixo:

| Variavel | Descrição|
|-|-|
|CRIM| taxa de crimes per-cápita da cidade | 
|ZN| proporção de terrenos residenciais zoneados para lotes com mais de 25.000 pés quadrados. |
|INDUS| proporção de acres de negócios não varejistas por cidade |
|CHAS |vale 1 se o terreno faz fronteira com o *Charles River*; 0 caso contrário |
|NOX| Concentração de óxido nítrico (em partes por 10 milhões) |
|RM| número médio de quartos por habitação |
|AGE| proporção de unidades ocupadas pelo proprietário construídas antes de 1940 |
|DIS| distâncias ponderadas até cinco centros de empregos de Boston |
|RAD| índice de acessibilidade a rodovias radiais |
|TAX| taxa de imposto sobre a propriedade de valor total por \\$10,000 |
|PTRATIO| razão pupilo-professor da cidade |
|B| $ 1000 (Bk - 0,63) ^ 2 $ onde Bk é a proporção de negros por cidade |
|LSTAT| \%status inferior da população |
|MEDV| (variável resposta) Valor mediano das casas ocupadas pelo proprietário em US $ 1.000|

````python
import graphviz

import pandas as pd

import seaborn as sns

from sklearn import datasets
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
from sklearn.model_selection import train_test_split


boston = datasets.load_boston()
X = pd.DataFrame(boston.data, columns = boston.feature_names)
y = pd.DataFrame(boston.target, columns = ['MEDV'])
````

````python
X.head()
````

````python
y.head()
````

### 1. Certifique-se de que esta base está no formato adequado para o scikitlearn.
ok, essa tarefa é tão fácil que você vai até desconfiar. Mas é preciso ter confiança sobre os dados que se usa ;)

````python
# Primeiramente temos que agrupar os 2 Dataframes em um

boston_raw = pd.concat([X, y], axis=1)
boston_raw.head()
````

````python
boston_raw.info()
````

Todos os dados já se encontram em formato quantitativo.

### 2.  Visualize a matriz de correlação, e veja se aparentemente você encontra alguma variável com potencial preditivo interessante.

````python
boston_raw.corr()
````

````python
cmap = sns.diverging_palette(h_neg=125, h_pos=350, as_cmap=True, sep = 60, center = 'light')

sns.clustermap(boston_raw.corr(), figsize=(10, 10), center = 0, cmap=cmap)
````

Acredito que existem duas variáveis com potencial preditivo interessante: **LSTAT** com corr de -0.737663 e **RM** com corr de 0.695360.

````python
# Como escolhi essas duas variáveis de análise criarei um novo Dataframe apenas com as variáveis de interesse

boston_a = boston_raw[['RM', 'LSTAT', 'MEDV']].copy()
boston_a.head()
````

### 3. Separe os dados em validação e teste

````python
X_a = boston_a.drop(columns = ['MEDV']).copy()
X_a.head()
````

````python
y_a = boston_a.loc[:,'MEDV']
y_a.head()
````

````python
X_train, X_test, y_train, y_test = train_test_split(X_a, y_a, random_state=2360873)
````

### 4. Treine duas árvores, uma com profundidade máxima = 8, outra com profundidade máxima = 2.

````python
# Decidi adicinar um min_samples_leaf = 10 para refinar a análise

regr_a = DecisionTreeRegressor(max_depth=8, min_samples_leaf=10)
regr_b = DecisionTreeRegressor(max_depth=2, min_samples_leaf=10)

regr_a.fit(X_train, y_train)
regr_b.fit(X_train, y_train)
````

### 5. Calcule o MSE de cada uma das árvores do item anterior para a base de treinamento e para a base de testes.

````python
# Para a base de treinamento

mse_a = regr_a.score(X_train, y_train)
mse_b = regr_b.score(X_train, y_train)

template = "O MSE da árvore com profundidade={0} é: {1:.2f}"

print(template.format(regr_a.get_depth(),mse_a).replace(".",","))
print(template.format(regr_b.get_depth(),mse_b).replace(".",","))
````

````python
# Para a base de teste

mse_c = regr_a.score(X_test, y_test)
mse_d = regr_b.score(X_test, y_test)

template = "O MSE da árvore com profundidade={0} é: {1:.2f}"

print(template.format(regr_a.get_depth(),mse_c).replace(".",","))
print(template.format(regr_b.get_depth(),mse_d).replace(".",","))
````

### 6. Com base nos resultados do item anterior, qual árvore te parece mais adequada?

A árvore com profundidade 8 parece muito interessante pois o MSE está mais próximo de 1. Porém a árvore com profundidade 2 também oferece um MSE alto com baixa profundidade, o que permite um modelo com menos overfitting e melhor visualização na árvore. Acredito que ambas sejam adequadas, mas vejo que para uma modelagem a de profundidade 2 seja mais útil, então optarei pela de profundidade 2 para o próximo exercício.
