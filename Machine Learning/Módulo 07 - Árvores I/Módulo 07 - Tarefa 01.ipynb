````python
# Criar um dataframe com os nomes de cada variável e o tipo de dados de cada variável.

col = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'qtd_filhos', 'tipo_renda', 'educacao', 
       'estado_civil', 'tipo_residencia', 'idade', 'tempo_emprego', 'possui_celular', 
       'possui_fone_comercial', 'possui_fone', 'possui_email', 'qt_pessoas_residencia', 'mau']

# Criando as colunas

demo_col_type = {}
demo_num_cat = {}

for i in col:
    demo_col_type[i] = demo[i].dtype
    demo_num_cat[i] = demo[i].nunique()
````

````python
demo_cat_a = pd.Series(demo_col_type)
demo_a = pd.DataFrame({'TIPO_DADOS': demo_cat_a})

demo_cat_b = pd.Series(demo_num_cat)
demo_b = pd.DataFrame({'NUM_CATEG': demo_cat_b})
````

````python
pd.concat([demo_a, demo_b], axis = 1)
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
# Variável "sexo"
demo.sexo.unique()
````

````python
demo.sexo = demo.sexo.map({'M':0, 'F':1 })
````

````python
# Variável "posse_de_veiculo"
demo.posse_de_veiculo.unique()
````

````python
demo.posse_de_veiculo = demo.posse_de_veiculo.map({'Y': 1, 'N': 0})
````

````python
# Variável "posse_de_imovel"
demo.posse_de_imovel.unique()
````

````python
demo.posse_de_imovel = demo.posse_de_imovel.map({'Y': 1, 'N': 0})
````

````python
# Criando variáveis Dummies para as variáveis "tipo_renda", "educacao", "estado_civil", "tipo_residencia"

demo = pd.get_dummies(demo, columns=['tipo_renda','educacao', 'estado_civil', 'tipo_residencia'], drop_first=True)
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
tab_a = pd.crosstab(demo['possui_email'], demo['mau'])
tab_a.div(tab_a.sum(axis=1), axis=0)
````

````python
tab_b = pd.crosstab(demo['posse_de_veiculo'], demo['mau'])
tab_b.div(tab_b.sum(axis=1), axis=0)
````

Apesar das porcentagens serem muito próximas podemos tirar algumas conclusões, se um cliente possui e-mail a chance dele ser inadimplente é menor comparado ao de um cliente que possui veiculo. Então podemos concluir que a variável "possui_email" é mais poderosa.
