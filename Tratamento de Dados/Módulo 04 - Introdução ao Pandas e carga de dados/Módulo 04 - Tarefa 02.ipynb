## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# Compilando os dados em um único DataFrame.

df_dados_notas = pd.DataFrame({'acertos_mat': df_mat['Qt_acertos'],
                           'acertos_por': df_por['Qt_acertos'],
                           'acertos_geo': df_geo['Qt_acertos'],
                           'acertos_ing': df_ing['Qt_acertos'],
                           'acertos_his': df_his['Qt_acertos'],
                           'acertos_fis': df_fis['Qt_acertos'],
                           'acertos_qui': df_qui['Qt_acertos']})
df_dados_notas.head()
````

````python
# 1) Calculo do percentual de acerto dos alunos por assunto e adicionar os valores no DataFrame original.

df_dados_notas.insert(1, 'porcentagem_mat', df_dados_notas['acertos_mat']/24)
df_dados_notas.insert(3, 'porcentagem_por', df_dados_notas['acertos_por']/18)
df_dados_notas.insert(5, 'porcentagem_geo', df_dados_notas['acertos_geo']/8)
df_dados_notas.insert(7, 'porcentagem_ing', df_dados_notas['acertos_ing']/8)
df_dados_notas.insert(9, 'porcentagem_his', df_dados_notas['acertos_his']/8)
df_dados_notas.insert(11, 'porcentagem_fis', df_dados_notas['acertos_fis']/12)
df_dados_notas.insert(13, 'porcentagem_qui', df_dados_notas['acertos_qui']/12)

df_dados_notas.head()
````

````python
# 2) Calculo do total de acertos de cada aluno e inserindo o valor na última coluna do Dataframe.

df_dados_notas['total_acertos'] = (df_dados_notas['acertos_mat'] +
                                  df_dados_notas['acertos_por'] +
                                  df_dados_notas['acertos_geo'] +
                                  df_dados_notas['acertos_ing'] +
                                  df_dados_notas['acertos_his'] +
                                  df_dados_notas['acertos_fis'] +
                                  df_dados_notas['acertos_qui'])

df_dados_notas['total_acertos']
````

````python
# 3) Calculo do porcentual geral de cada aluno e inserção deste na última coluna.
# Temos um total de 90 questões

df_dados_notas['porcentagem_total_acertos'] = (df_dados_notas['total_acertos'] / 90)

df_dados_notas['porcentagem_total_acertos']
````

````python
# 4) Calculo de quantos alunos tiveram nota maior que 45.
# Notas válidas são notas acima de 45 (nota válida > 45)

df_dados_notas['aprovados'] = 0

df_dados_notas.loc[df_dados_notas['total_acertos'] > 45, 'aprovados'] = 1

df_dados_notas['aprovados'].value_counts()
````

````python
# 4) Continuação

qtd_aprovados = df_dados_notas['aprovados'].sum()
print(f'A quantidade de aprovados é: {qtd_aprovados}.')
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# 1) Identificando alunos desclassificados

df_dados_notas['desclassificados'] = 0

df_dados_notas.loc[(df_dados_notas['acertos_mat'] == 0) | (df_dados_notas['acertos_fis'] == 0) | (df_dados_notas['acertos_qui'] == 0), 'desclassificados'] = 1
df_dados_notas['desclassificados']
````

````python
# 1) Criando um novo DataFrame apenas para os desclassificados
df_dados_notas_desc = df_dados_notas[df_dados_notas['desclassificados'] == 1]
df_dados_notas_desc
````

````python
# 2) Cálculo dos alunos desclassificados

qtd_desclassificados = df_dados_notas['desclassificados'].sum()
print(f'A quantidade de desclassificados é: {qtd_desclassificados}.')
````

````python
# 3) Média, em história e geografia, dos alunos desclassificados alunos.

media_geo_desc = df_dados_notas_desc['acertos_geo'].mean()
print(f'A média de acertos em geografia dos alunos desclassificados é: {media_geo_desc}')
media_his_desc = df_dados_notas_desc['acertos_his'].mean()
print(f'A média de acertos em história dos alunos desclassificados é: {media_his_desc}')
````

````python
# 4) Dataframe para os aprovados e não desclassificados

df_aprovados = df_dados_notas[(df_dados_notas['desclassificados'] == 0) & (df_dados_notas['aprovados'] == 1)]
df_aprovados
````
